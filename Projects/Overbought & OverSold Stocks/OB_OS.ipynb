Aim: To create a dashboard to get the last day Overbought & Oversold Stocks on the basis of RSI value from Top 500 Companies

```python
import pandas as pd
import yfinance as yf
#from nsetools import Nse
from datetime import datetime, timedelta
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
from io import BytesIO
import base64

print("libraries imported")
```

```python
# Importing Data of all the stocks (as on 30-Dec'22)
stock_list_df = pd.read_excel("C:\\Users\\rahul.vimal\\Documents\\Python\\Personal_Work\\Stocks\\NSE Stocks.xlsx")
stock_list_df.head()
```

```python
# Define the number of days to go back
days_to_go_back = 25

# Calculate the start date as today's date minus the specified number of days
end_date = datetime.today().date()
start_date = end_date - timedelta(days=days_to_go_back)

# Filtering Top 500 Stocks
top_500_stocks = stock_list_df.iloc[:500][:]

# Extracting data for the last 20 days from today's date
stocks_data_raw = pd.DataFrame()

#nse = Nse()
#nse_live = NSELive() # Initializing jugaad_data.nse live

k = 1
for i in top_500_stocks.symbol:
    print(k)
    k += 1
    hist_data = yf.download(i + ".NS", start=start_date, end=end_date) # Extracting Stock Data
    hist_data['symbol'] = i # Adding Stock Name
    stocks_data_raw = pd.concat([stocks_data_raw, hist_data], axis=0) # Appending Dataframes

stocks_data = stocks_data_raw[['Close', 'symbol']]
stocks_data = stocks_data.sort_values(by=['symbol', 'Date'], ascending=True) # Sort the data in descending order
stocks_data.reset_index(inplace=True)
```

```python
# Define the number of periods for RSI calculation
rsi_period = 14  # Adjust as needed

# Calculate the daily price changes
stocks_data['Price Change'] = stocks_data.groupby('symbol')['Close'].diff()

# Separate gains (positive price changes) and losses (negative price changes)
stocks_data['Gain'] = stocks_data['Price Change'].apply(lambda x: x if x > 0 else 0)
stocks_data['Loss'] = stocks_data['Price Change'].apply(lambda x: -x if x < 0 else 0)

# Calculate the average gain and average loss over the RSI period
stocks_data['Avg Gain'] = stocks_data.groupby('symbol')['Gain'].rolling(window=rsi_period, min_periods=1).mean().reset_index(0, drop=True)
stocks_data['Avg Loss'] = stocks_data.groupby('symbol')['Loss'].rolling(window=rsi_period, min_periods=1).mean().reset_index(0, drop=True)

# Calculate the Relative Strength (RS)
stocks_data['RS'] = stocks_data['Avg Gain'] / stocks_data['Avg Loss']

# Calculate the RSI
stocks_data['RSI'] = 100 - (100 / (1 + stocks_data['RS']))

stocks_data.to_csv('Sample_output.csv',index=False)

# Group the data by symbol and extract the last RSI value in each group
#last_rsi_values = stocks_data.groupby('symbol')['RSI'].last().reset_index()
last_rsi_values = stocks_data.groupby('symbol')[['RSI','Close']].last().reset_index()

last_rsi_values.to_csv('RSI_output.csv',index=False)

print("Output Successfully exported")
```

```python


########################################
OB_OS = last_rsi_values.query('RSI < 30 or RSI > 70').sort_values('RSI',ascending=False)
OB_OS.loc[OB_OS['RSI'] < 30,'Type'] = 'OverSold'
OB_OS.loc[OB_OS['RSI'] > 70,'Type'] = 'OverBought'
OB_OS['Date'] = (datetime.now() - timedelta(days=1)).strftime('%d-%b-%y')




# Sample DataFrame (Replace this with your actual DataFrame)
df = OB_OS[['Date','Type','symbol','RSI','Close']].copy()

# Email content and details
mail_content = '''Hello,
This is a simple mail with the list of stocks which were either OverSold or Overbought Yesterday.
Thank You
'''

sender_address = 'xxxxxxxx@gmail.com'
sender_pass = 'xxxxxxxx'  # Your Gmail app password
receiver_address = 'xxxxxxxx@gmail.com'

# Calculate the date for yesterday
yesterday = datetime.now() - timedelta(days=1)

# Format yesterday's date in "DD-MMM-YY" format
yesterday_date = yesterday.strftime('%d-%b-%y')

# Set the email subject with the current date
email_subject = f'Overbought & Oversold Stocks List on ({yesterday_date})'

# Setup the MIME
message = MIMEMultipart()
message['From'] = sender_address
message['To'] = receiver_address
message['Subject'] = email_subject

# Attach the email body
message.attach(MIMEText(mail_content, 'plain'))

# Create an HTML table from the DataFrame with color formatting
def color_format(row):
    if row['Type'] == 'OverBought':
        return f'<span style="color: red; font-weight: bold;">{row["Type"]}</span>'
    elif row['Type'] == 'OverSold':
        return f'<span style="color: green; font-weight: bold;">{row["Type"]}</span>'
    else:
        return row['Type']

df['Type'] = df.apply(color_format, axis=1)
html_table = df.to_html(index=False, escape=False)  # Disable HTML escaping

# Attach the HTML table in the email body
html_part = MIMEText(html_table, 'html')
message.attach(html_part)

# Convert DataFrame to CSV and attach it as an attachment
csv_string = df.to_csv(index=False)  # Convert DataFrame to CSV string
csv_part = MIMEBase('application', 'octet-stream')
csv_part.set_payload(csv_string)
encoders.encode_base64(csv_part)
csv_part.add_header('Content-Disposition', 'attachment; filename="sample_data.csv"')
message.attach(csv_part)

# Create SMTP session for sending the mail
session = smtplib.SMTP('smtp.gmail.com', 587)  # use Gmail with port
session.starttls()  # enable security
session.login(sender_address, sender_pass)  # login with mail_id and password
text = message.as_string()
session.sendmail(sender_address, receiver_address, text)
session.quit()

print('Mail Sent')

```
